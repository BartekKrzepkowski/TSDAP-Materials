In general, database normalization is the process of structuring a database to reduce data redundancy. (see e.g. https://en.wikipedia.org/wiki/Database_normalization)

For a given dataframe mleaders perform below operations to normalize database.

1) Goal: The values in each column of a table must be atomic. HINT: Replace a row of data of a leader with two battles with two rows, one per each battle.

2) Goal: Remove Transitive Dependency, that is, situation when a non-prime attribute depends on other non-prime attributes rather than depending upon the prime attributes or primary key. (more info: https://en.wikipedia.org/wiki/Transitive_dependency) HINT: Divide data in two tables - battle, leaders. HINT: remove duplicate rows.

3) Goal: Remove Partial Dependency, which exists, when for a composite primary key, any attribute in the table depends only on a part of the primary key and not on the complete primary key. HINT: The key for leaders table consists of columns 'Leader Name' and 'Leader Surame'

4) Sum DataFrames from the result of part 3) to recreate table leaders obtained from part 2). HINT: use join

1), 2) and 3) should simulate 1NF, 3NF, 2NF.
