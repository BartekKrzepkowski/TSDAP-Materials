```python
from skimage import io
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
```

Download an image from https://i.stack.imgur.com/VxtPG.png
To show pictures you can use pyplot from matplotlib.
See functions:
imread https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imread.html,
imshow https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html,
show https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html.

Your goal is to make the picture more "smooth" using Gaussian filter https://en.wikipedia.org/wiki/Gaussian_filter.
Fortunately, it is already implemented - scipy.ndimage.gaussian_filter (https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.gaussian_filter.html)
Above function has two non-optional parameters: (a) input (which is an array of numbers derived from image), and
(b) sigma (which impact on smoothness level)

You should use the argparse in your script.
Usage: myscript.py [-h] [-x XVAL] [-y YVAL] [-xy VAL] inputfile

Your script should display image from inputfile. Apply gaussian_filter(input, sigma), and display modified image.
If user do not specify optional arguments, default value of sigma should be 2.
If user specifies -x 1 -y 4, the sigma should be (1,4).
If user specifies -xy 3, the sigma should be 3.`

```python
url = 'https://i.stack.imgur.com/VxtPG.png'
image = io.imread(url)
plt.figure(figsize = (10,10))
plt.imshow(image)
plt.show()
```

```python
filtered_img = gaussian_filter(image, 1)
plt.figure(figsize = (10,10))
plt.imshow(filtered_img)
plt.show()
```
